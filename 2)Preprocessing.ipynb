```python
import pandas as pd
import numpy as np
import eda
import preprocess

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline

import warnings
warnings.filterwarnings('ignore')

pd.pandas.set_option('display.max_columns',None)
```

```python
# Reading in the trainset.
data = pd.read_csv('train.csv')
data.head()
```

```python
train,test = train_test_split(data, test_size=0.15)
```

# Creating/Removing/Transforming the features

## Historic variables

```python
eda.historic_data(train).head()
```

In the EDA stage we found that the new features indicating the age of the building and its aspects have a relationship with the price. So we create those features and replace the original features with the new ones.

## Log-Transformation of the continous variables

In the EDA stage we found that, we can normalize some of the features by applying a log-transformation on them. So we implement a function to perform this transformation.

```python
preprocess.log_normalize(train)[eda.numeric_continous(train).columns].head()
    
```

## Dealing with the rare categories of the categorical features

We try to ignore the categories of a categorical feature which constitute a very small fraction of that feature. For instance, if a class only fills in less than 1% of the observations within a feature, we will ignore it.

```python
preprocess.rare_categories(train).head()
```

# Converting the categorical features to numeric

To apply a machine learning algorithm, we need to convert the categorical features to numeric.

```python
preprocess.cat_to_numeric(train).head()
```

# Dropping unwanted features

```python
preprocess.drop_feature(train).head(2)
```

# Pipeline

Having defined all the requiered functions, we are now able to create a preparation pipeline to transform the raw data into an appropriate dataset to get ready for modeling.

```python
preprocess_pipe = Pipeline(steps=[('cat_impute', FunctionTransformer(preprocess.impute_cat_features)),
                                  ('numeric_impute',FunctionTransformer(preprocess.impute_numeric_data)),
                                  ('age_feature',FunctionTransformer(preprocess.create_age_feature)),
                                  ('log_transform',FunctionTransformer(preprocess.log_normalize)),
                                  ('rare_class', FunctionTransformer(preprocess.rare_categories)),
                                  ('cat_to_numeric', FunctionTransformer(preprocess.cat_to_numeric)),
                                  ('drop_feature',FunctionTransformer(preprocess.drop_feature) )
                                 ])
```

```python
preprocess = preprocess_pipe.fit(train)
train_set = preprocess.transform(train)
test_set = preprocess.transform(test)
```
