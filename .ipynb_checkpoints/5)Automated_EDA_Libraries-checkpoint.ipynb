## 2) [Pandas-Profiling](https://pandas-profiling.github.io/pandas-profiling/docs/master/rtd/#)

```python
df1 = sns.load_dataset('tips')
df1.shape
```

```python
#!pip install pandas-profiling[notebook]
from pandas_profiling import ProfileReport
profile = ProfileReport(df1, title="tips Report", explorative=True)
```

```python

profile.to_file('output.html')
```

```python
profile.to_widgets()
```

## 3) [sweetviz](https://pypi.org/project/sweetviz/)

```python
#!pip install sweetviz
```

```python
import sweetviz as sv
report = sv.analyze(df)
report.show_html('Titanic.html')
```

## 4) [Autoviz](https://pypi.org/project/autoviz/)

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
```

```python
filename = ""
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=df,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
